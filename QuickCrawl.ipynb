```python
import tweepy as tp
from cred import *
import pandas as pd     # To handle data
import numpy as np      # For number computing
from IPython.display import display
```

```python
def setTwitterAuth():
    """
    obtains authorization from twitter API
    """
    auth = tp.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token_key, access_token_secret)
    api = tp.API(auth)
    return api

api = setTwitterAuth()
user = api.me()
print 'Using account: @' + user.screen_name
```

```python
def searchTweet(api, searchTerm, searchlim):
    """
    gets search results of the string, search, and returns them as a list
    of tweet objects
    """
    tweets = [status for status in tp.Cursor(api.search, q=searchTerm).items(searchlim)]
    return tweets
```

```python
print 'Crawling'
searchlim = 10
searchterm = '"fake news" AND "detection"-filter:retweets '
tweets = searchTweet(api,searchterm,searchlim)

#tweets = GetUserTimeline(api,'realDonaldTrump ',searchlim)
#for tweet in tweets:
    #print tweet.text + '\n'
    #print tweet.user.screen_name


# Internal methods of a single tweet object:
# print(dir(tweets[0]))


print 'Done: ' + str(len(tweets)) + ' tweets found'
```
